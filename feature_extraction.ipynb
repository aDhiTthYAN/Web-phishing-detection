```python
def is_indexed_by_google(url):
    try:
        response = requests.get(f"https://www.google.com/search?q=site:{url}")
        if response.status_code == 200:
            if "did not match any documents" in response.text:
                return 0
            else:
                return 1
        else:
            return -1
    except Exception as e:
        print(f"Error rettrieving Google search reuslts: {str(e)}")
        return -1
```

```python
legit["indexed_by_google"] = legit["URL"].apply(is_indexed_by_google)
```
