<h3>Length of URL</h3>

*  Computes the length of the URL. Phishers can use long URL to hide the doubtful part in the address bar. In this project, if the length of the URL is    greater than or equal 54 characters then the URL classified as phishing otherwise legitimate.
*  If the length of URL >= 54 , the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
def get_url_length(url):
    if len(url) >= 54:
        return 1
    else:
        return 0
```

<h3>Prefix or Suffix "-" in Domain</h3>

*  Checking the presence of '-' in the domain part of URL. The dash symbol is rarely used in legitimate URLs. Phishers tend to add prefixes or suffixes    separated by (-) to the domain name so that users feel that they are dealing with a legitimate webpage.
*  If the URL has '-' symbol in the domain part of the URL, the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
def has_hyphens(url):
    if '-' in urlparse(url).netloc: # we will be using the urlparse from urllib.parse for checking the presence of '-' in the url
        return 1
    else:
        return 0
```

```python
legit["has_hyphens"] = legit["URL"].apply(has_hyphens)
```

```python
def has_http_in_domain(url):
    parsed_url = urlparse(url)
    domain = parsed_url.netloc.split(':')[0]
    if 'http' in domain:
        return 1
    else:
        return 0
```

```python
import asyncio
import aiohttp
import pandas as pd

async def fetch_website_traffic(url):
    async with aiohttp.ClientSession() as session:
        try:
            async with session.get(f"https://awis.api.alexa.com/api?Action=TrafficHistory&Range=31&ResponseGroup=History&Url={url}&Output=json") as response:
                data = await response.json()
                return url, data['TrafficHistory']
        except Exception as e:
            print(f"Error fetching traffic for {url}: {e}")
            return url, None

async def fetch_website_traffic_for_batch(urls):
    tasks = []
    async with aiohttp.ClientSession() as session:
        for url in urls:
            tasks.append(fetch_website_traffic(url))
        results = await asyncio.gather(*tasks)
        return results

async def main():
    phish_urls = phish['URL'].tolist()  # Extract URLs from DataFrame
    legit_urls = legit['URL'].tolist()  # Extract URLs from DataFrame

    # Define batch size for each DataFrame
    phish_batch_size = 5
    legit_batch_size = 5

    # Fetch website traffic for phishing URLs
    for i in range(0, len(phish_urls), phish_batch_size):
        batch_urls = phish_urls[i:i+phish_batch_size]
        results = await fetch_website_traffic_for_batch(batch_urls)
        for url, traffic in results:
            if traffic is not None:
                print(f"Traffic for phishing URL {url}: {traffic}")
            else:
                print(f"Failed to fetch traffic for phishing URL {url}")

    # Fetch website traffic for legitimate URLs
    for i in range(0, len(legit_urls), legit_batch_size):
        batch_urls = legit_urls[i:i+legit_batch_size]
        results = await fetch_website_traffic_for_batch(batch_urls)
        for url, traffic in results:
            if traffic is not None:
                print(f"Traffic for legitimate URL {url}: {traffic}")
            else:
                print(f"Failed to fetch traffic for legitimate URL {url}")
```

```python
import asyncio
import aiohttp
import pandas as pd

async def fetch_alexa_rank(url):
    async with aiohttp.ClientSession() as session:
        try:
            async with session.get(f"https://awis.api.alexa.com/api?Action=UrlInfo&Url={url}&Output=json") as response:
                data = await response.json()
                return url, data['UrlInfo']['Rank']
        except Exception as e:
            print(f"Error fetching Alexa rank for {url}: {e}")
            return url, None

async def fetch_alexa_rank_for_batch(urls):
    tasks = []
    async with aiohttp.ClientSession() as session:
        for url in urls:
            tasks.append(fetch_alexa_rank(url))
        results = await asyncio.gather(*tasks)
        return results

async def main():
    phish_urls = phish['URL'].tolist()  # Extract URLs from DataFrame
    legit_urls = legit['URL'].tolist()  # Extract URLs from DataFrame

    # Define batch size for each DataFrame
    phish_batch_size = 5
    legit_batch_size = 5

    # Fetch Alexa rank for phishing URLs
    phish_rank = []
    for i in range(0, len(phish_urls), phish_batch_size):
        batch_urls = phish_urls[i:i+phish_batch_size]
        results = await fetch_alexa_rank_for_batch(batch_urls)
        phish_rank.extend(results)

    # Fetch Alexa rank for legitimate URLs
    legit_rank = []
    for i in range(0, len(legit_urls), legit_batch_size):
        batch_urls = legit_urls[i:i+legit_batch_size]
        results = await fetch_alexa_rank_for_batch(batch_urls)
        legit_rank.extend(results)

    # Add Alexa rank data to DataFrame
    phish_rank_df = pd.DataFrame(phish_rank, columns=['URL', 'Rank'])
    phish = pd.merge(phish, phish_rank_df, on='URL', how='left')
    phish['Alexa_Rank'] = phish['Rank'].apply(lambda x: 1 if x is not None and x < 100000 else 0)

    legit_rank_df = pd.DataFrame(legit_rank, columns=['URL', 'Rank'])
    legit = pd.merge(legit, legit_rank_df, on='URL', how='left')
    legit['Alexa_Rank'] = legit['Rank'].apply(lambda x: 1 if x is not None and x < 100000 else 0)

    # Save DataFrame with Alexa rank data to CSV files
    phish.to_csv('structured_data_phishing_with_rank.csv', index=False)
    legit.to_csv('structured_data_legitimate_with_rank.csv', index=False)
```
