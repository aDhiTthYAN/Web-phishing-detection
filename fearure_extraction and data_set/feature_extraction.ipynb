<h3>Length of URL</h3>

*  Computes the length of the URL. Phishers can use long URL to hide the doubtful part in the address bar. In this project, if the length of the URL is    greater than or equal 54 characters then the URL classified as phishing otherwise legitimate.
*  If the length of URL >= 54 , the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
def get_url_length(url):
    if len(url) >= 54:
        return 1
    else:
        return 0
```

<h3>Prefix or Suffix "-" in Domain</h3>

*  Checking the presence of '-' in the domain part of URL. The dash symbol is rarely used in legitimate URLs. Phishers tend to add prefixes or suffixes    separated by (-) to the domain name so that users feel that they are dealing with a legitimate webpage.
*  If the URL has '-' symbol in the domain part of the URL, the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
def has_hyphens(url):
    if '-' in urlparse(url).netloc: # we will be using the urlparse from urllib.parse for checking the presence of '-' in the url
        return 1
    else:
        return 0
```

```python
legit["has_hyphens"] = legit["URL"].apply(has_hyphens)
```

```python
phish["has_hyphens"] = phish["URL"].apply(has_hyphens)
```

<h3> IP Address in the URL</h3>

*  Checks for the presence of IP address in the URL. URLs may have IP address instead of domain     name. If an IP address is used as an alternative of the domain name in the URL, we can be sure    that someone is trying to steal personal information with this URL.
*  If the domain part of URL has IP address, the value assigned to this feature is 1 (phishing)      or else 0 (legitimate).

```python
import ipaddress # imported ip address module for getting the ip out of the url

def having_ip(url):
    try:
        parsed_url = urlparse(url)
        ipaddress.ip_address(parsed_url.netloc)#using ipaddress.ip_address we will be checking whether the url have the ip address in it or not
        return 1
    except ValueError:
        return 0
```

```python
phish["having_ip"] = phish["URL"].apply(having_ip)
```

```python
phish["having_ip"].value_counts()
```

```python
legit["having_ip"] = legit["URL"].apply(having_ip)
```

<h3> "@" Symbol in URL</h3>

*   Checks for the presence of '@' symbol in the URL. Using “@” symbol in the URL leads the browser to ignore everything preceding the “@” symbol and      the real address often follows the “@” symbol.
*   If the URL has '@' symbol, the value assigned to this feature is 1 (phishing) or else 0 (legitimate).

```python
def having_atsign(url):
    if '@' in urlparse(url).netloc:
        return 1
    else:
        return 0
```

<h3>Subdomain Count</h3>


```python
import tldextract
def count_subdomains(url):
    # Use tldextract library to extract subdomains
    ext = tldextract.extract(url)
    subdomains = ext.subdomain.split('.')  # Split subdomains by dot separator
    # Return the count of subdomains
    return len(subdomains)
```

```python
phish['subdomain_count'] = phish['URL'].apply(count_subdomains)
```

<h3>
Top-Level Domain (TLD) Type</h3

* Identify the type of TLD used (e.g.,top,info .org, .xyz, etc.). Certain TLDs are more commonly associated with phishing.>

```python
def mark_phishing_tld(url):
    ext = tldextract.extract(url)
    phishing_tlds = ['xyz', 'top', 'info', 'loan', 'xmr', 'ink', 'sk', 'ag', 'ryukyu', 'ltc', 'gr', 'fund', 'sx']

    # Convert both TLD and phishing_tlds list to lowercase for case-insensitive comparison
    tld = ext.suffix.lower()
    phishing_tlds = [tld.lower() for tld in phishing_tlds]

    if tld in phishing_tlds:
        return 1  # Mark as phishing
    else:
        return 0  # Mark as legitimate
```

```python
phish['Phishing_TLD'] = phish['URL'].apply(mark_phishing_tld)
```

```python
phish['Phishing_TLD'].value_counts()
```

```python
legit['Phishing_TLD'] = legit['URL'].apply(mark_phishing_tld)
```
